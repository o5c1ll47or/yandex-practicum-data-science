## Описание проекта

Вы аналитик компании «Мегалайн» — федерального оператора сотовой связи. Клиентам предлагают два тарифных плана: «Смарт» и «Ультра». Чтобы скорректировать рекламный бюджет, коммерческий департамент хочет понять, какой тариф приносит больше денег.
Вам предстоит сделать предварительный анализ тарифов на небольшой выборке клиентов. В вашем распоряжении данные 500 пользователей «Мегалайна»: кто они, откуда, каким тарифом пользуются, сколько звонков и сообщений каждый отправил за 2018 год. Нужно проанализировать поведение клиентов и сделать вывод — какой тариф лучше.

**Описание тарифов**

***Тариф «Смарт»***
1. Ежемесячная плата: 550 рублей
2. Включено 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика
3. Стоимость услуг сверх тарифного пакета:
    * минута разговора: 3 рубля
    * сообщение: 3 рубля
    * 1 Гб интернет-трафика: 200 рублей

***Тариф «Ультра»***
1. Ежемесячная плата: 1950 рублей
2. Включено 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика
3. Стоимость услуг сверх тарифного пакета:
    * минута разговора: 1 рубль
    * сообщение: 1 рубль
    * 1 Гб интернет-трафика: 150 рублей

**Обратите внимание**: «Мегалайн» всегда округляет вверх значения минут и мегабайтов. Если пользователь проговорил всего 1 секунду, в тарифе засчитывается целая минута.

1. Таблица **users** (информация о пользователях):

    * `tariff` — название тарифного плана


2. Таблица **calls** (информация о звонках):
    * `id` — уникальный номер звонка
    * `call_date` — дата звонка

    * `user_id` — идентификатор пользователя, сделавшего звонок


3. Таблица **messages** (информация о сообщениях):
    * `id` — уникальный номер сообщения
    * `message_date` — дата сообщения

5. Таблица **tariffs** (информация о тарифах):

    * `rub_per_message` — стоимость отправки сообщения сверх тарифного пакета
    * `rub_per_gb` — стоимость дополнительного гигабайта интернет-трафика сверх тарифного пакета (1 гигабайт = 1024 мегабайта)

### Приведение данные к нужным типам


Применим к столбцам с датами `pd.to_datetime`:

```python
df_users['reg_date'] = pd.to_datetime(df_users['reg_date'], format='%Y-%m-%d')
df_users['churn_date'] = pd.to_datetime(df_users['churn_date'], format='%Y-%m-%d')
df_calls['call_date'] = pd.to_datetime(df_calls['call_date'], format='%Y-%m-%d')
df_messages['message_date'] = pd.to_datetime(df_messages['message_date'], format='%Y-%m-%d')
df_internet['session_date'] = pd.to_datetime(df_internet['session_date'], format='%Y-%m-%d')
```

### Посчитаем для каждого пользователя:

Для этого воспользуемся `pivot_table` и переименованием столбцов. Выведем первые пять строк каждой таблицы:

#### количество сделанных звонков и израсходованных минут разговора по месяцам;

```python
id_calls = df_calls.pivot_table(index = ['user_id','month'], values = 'duration', aggfunc = ('count', 'sum'))
id_calls.columns = ['call_counts', 'minutes_spent']
id_calls.head()
```

#### количество отправленных сообщений по месяцам;

```python
id_messages = df_messages.pivot_table(index = ['user_id','month'], values = 'id', aggfunc = ('count'))
id_messages.columns = ['messages_spent']
id_messages.head()
```

#### объем израсходованного интернет-трафика по месяцам;

```python
id_internet = df_internet.pivot_table(index = ['user_id', 'month'], values = 'mb_used', aggfunc = ('sum'))
id_internet.columns = ['mb_spent']
id_internet.head()
```

#### помесячную выручку с каждого пользователя

```python
df=id_calls.merge(id_messages,on=['user_id','month'],how='outer')
df=df.merge(id_internet,on=['user_id','month'],how='outer')
df.head()
```

```python
def proceeds(x):
    minutes_spent = x['minutes_spent']
    messages_spent = x['messages_spent']
    mb_spent = x['mb_spent']
    tariff = x['tariff']
    
    if tariff == 'smart':
        i = 0
    else:
        i = 1
        
    extra_minutes = minutes_spent - df_tariffs.loc[i, 'minutes_included']
    extra_messages = messages_spent - df_tariffs.loc[i, 'messages_included']
    extra_mb = mb_spent - df_tariffs.loc[i, 'mb_per_month_included']
        
    if extra_minutes < 0:
        extra_minutes = 0
    if extra_messages < 0:
        extra_messages = 0
    if extra_mb < 0:
        extra_mb = 0
    
    return (df_tariffs.loc[i, 'rub_per_message'] * extra_messages + (df_tariffs.loc[i, 'rub_per_gb'] / 1024) * extra_mb + df_tariffs.loc[i, 'rub_per_minute'] * extra_minutes + df_tariffs.loc[i, 'rub_monthly_fee'])
```

Добавим с помощью написанной функции новый столбец. Выведем первые 5 строк:

```python
df['proceeds'] = df.apply(proceeds,axis=1)
df.head()
```

## Шаг 4. Проверим гипотезы

### Cредняя выручка пользователей тарифов «Ультра» и «Смарт» различается

Сформируем нуливую и альтернативную гипотизу:

$H_0 $ — средняя выручка пользователей тарифа «Ультра» равна средней выручке пользователей тарифа «Смарт».

$H_1 $ — средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются.

$α$ = 5%

<div class="alert alert-success">
    
<b> Комментарий от ревьюера</b>
    
Ты правильно сформулировал гипотезы.

```python
df_ultra = df_sum.query('tariff == "ultra"')['proceeds']
df_smart = df_sum.query('tariff == "smart"')['proceeds']

alpha = 0.05

results = st.ttest_ind(df_ultra, df_smart, equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```
