ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

```python
def life(data):
    life_lst=[]
    for platform in data.columns:
        year = data[platform].ne(0).value_counts().loc[data[platform].ne(0).value_counts().index == True].values[0]
        life_lst.append(year)
    return median(life_lst)
```

### Посмотрим, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Построим диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами. Сформулируем выводы.

Напишем функцию для построения диаграмм рассеяния:

```python
def corr(data, platform):
    platform_data = data[(data['platform'] == platform) & (data['critic_score'] > 0) & (data['user_score'] > 0)]
    s1 = figure(title = "Корреляция для {} (пользователи)".format(platform))
    s1.xaxis.axis_label = 'user_score'
    s1.yaxis.axis_label = 'sales_total'

    s1.circle(x='user_score', y='sales_total', size=5, alpha=0.5, source=platform_data)

    s2 = figure(title = "Корреляция для {} (критики)".format(platform))
    s2.xaxis.axis_label = 'critic_score'
    s2.yaxis.axis_label = 'sales_total'
    s2.circle(x='critic_score', y='sales_total', size=5, alpha=0.5, source=platform_data)

    s1.xgrid.grid_line_color = None
    s1.ygrid.grid_line_color = None
    s1.outline_line_color = None

    s2.xgrid.grid_line_color = None
    s2.ygrid.grid_line_color = None
    s2.outline_line_color = None

    grid = gridplot([[s1, s2]], plot_width=450, plot_height=250)

    output_notebook()

    show(grid)
```

Построим диаграмму для самой популярной платформы - **PS2**:

```python
corr(df, 'PS2')
```

На диаграммах можно наблюдать, что более высокие отзывы приводят к более высоким продажам. Отзывы критиков вносят немного больше вклада в продажи, чем пользователи. Рассчитаем коэффициенты корреляции:

```python
ps2 = df[(df['platform'] == 'PS2') & (df['critic_score'] > 0) & (df['user_score'] > 0)]
print('Корреляция между отзывами пользователей и продажами: {:.2f}'.format(ps2['user_score'].corr(ps2['sales_total'])))
print('Корреляция между отзывами критиков и продажами: {:.2f}'.format(ps2['critic_score'].corr(ps2['sales_total'])))
```

Действительно, корреляция для критиков выше, однако сами значения корреляции не столь велики.

## Шаг 4. Составим портрет пользователя каждого региона

Определим для пользователя каждого региона (NA, EU, JP). Напишем для построения графиков функцию:

```python
def regions(data, title=''):
    labels = data.index
    regions = ['eu_sales', 'jp_sales','na_sales']
    
    x = [ (label, region) for label in labels for region in regions ]
    counts = sum(zip(data['eu_sales'], data['jp_sales'], data['na_sales']), ())
    
    source = ColumnDataSource(data=dict(x=x, counts=counts))
    
    p = figure(x_range=FactorRange(*x),
           plot_width=900,
           plot_height=300, 
           title=title,
           toolbar_location=None,
           tools="",
           tooltips="Частота: @counts")
    
    p.vbar(x='x',
       top='counts',
       width=0.9,
       source=source,
       line_color="white",
       fill_color=factor_cmap('x', palette=['#F14040', '#1A6FDF','#37AD6B'], factors=regions, start=1, end=2))

    p.y_range.start = 0
    p.x_range.range_padding = 0.1
    p.xaxis.major_label_orientation = 1
    p.xgrid.grid_line_color = None
    p.xgrid.grid_line_color = None
    p.ygrid.grid_line_color = None
    p.outline_line_color = None

    output_notebook()
    show(p) 
```

<div class="alert alert-info" style="border-radius: 15px">

Не очень понял замечание, но предположу, что гипотезы необходимо было проверить именно на актуальном периоде. Добавил параллельно расчет.
    
Результаты полностью противоречат друг другу:
1. Во всем периоде отзывы выше у <b>PS</b>, однако в актуальном разницы нет.
2. Во всем периоде рейтинг одинаковый у жанров <b>Action</b> и <b>Sport</b>, однако в актуальном периоде рейтинги у <b>Action</b> выше.

</div>

#### Для всего периода

```python
xone = df.query('platform == "XOne" & user_score > 0')['user_score']
pc = df.query('platform == "PC" & user_score > 0')['user_score']

alpha = 0.05

results = st.ttest_ind(xone, pc, equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Посмотрим на среднее и доверительный интервал:

```python
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), st.sem(a)
    h = se * st.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h
```

```python
mean_confidence_interval(xone)
```

```python
mean_confidence_interval(pc)
```

Отзывы на **PC** в среднем на 0,5 выше чем на **Xbox One**.

#### Для актуального периода # new

```python
xone = df_actual.query('platform == "XOne" & user_score > 0')['user_score'] #new
pc = df_actual.query('platform == "PC" & user_score > 0')['user_score'] #new

alpha = 0.05 #new

results = st.ttest_ind(xone, pc, equal_var=False) #new

print('p-значение:', results.pvalue) #new

if (results.pvalue < alpha): #new
    print("Отвергаем нулевую гипотезу") #new
else: #new
    print("Не получилось отвергнуть нулевую гипотезу") #new
```

```python
mean_confidence_interval(xone)
```

Среднее и доверительный интервал практически совпадают.
